```python
# code from https://www.kaggle.com/kalilurrahman/olympic-games-eda to adjust historical country names to their modern names.

medals_cleaned["country_code"].replace({"URS": "RUS",
                             "ROC": "RUS",
                             "OAR": "RUS",
                             "ORS": "RUS",
                             "GDR": "GER",
                             "FRG": "GER"}, inplace=True)
medals_cleaned["country_name"].replace({"Russian Federation": "Russia",
                        "Soviet Union": "Russia",
                        "Unified team": "Russia",
                        "West Germany": "Germany",
                        "East Germany": "Germany"}, inplace=True)
```

```python
#Merge olympics_hosts and olympics_medals. Code based off eda from https://www.kaggle.com/kalilurrahman/olympic-games-eda

olympics_results = pd.merge(hosts_cleaned, medals_cleaned,
                           left_on=  ['game_slug'],
                           right_on= ['slug_game'],
                           how = 'left')
olympics_results.head(10)
```

```python
olympics_results.columns
```

```python
olympics_results.value_counts(['medal_type'])
```

```python
clean_olympics_results= olympics_results.drop(['game_end_date', 'game_start_date', 'game_name', 'game_season', 'game_year', 'discipline_title', 'game_slug', 'event_title', 'event_gender', 'medal_type', 'participant_type', 'country_3_letter_code', 'slug_game', 'country_code'], 1)
```

```python
clean_olympics_results.dropna()
```

```python
#making new df with rows where game_location & country_name match

matching = clean_olympics_results[clean_olympics_results['game_location'] == clean_olympics_results['country_name']]
matching
```

```python
# adding column of home field winners
clean_olympics_results['home_winner'] = np.where(clean_olympics_results['game_location'] == clean_olympics_results['country_name'], True, False)
clean_olympics_results.head()
```

```python
# Converting true/false into 1, 0 https://stackoverflow.com/questions/17383094/how-can-i-map-true-false-to-1-0-in-a-pandas-dataframe
clean_olympics_results["home_winner"] = clean_olympics_results["home_winner"].astype(bool).astype(int)
```
